```python
import pandas as pd
%matplotlib inline 
```

```python
# Lendo o arquivo.
df = pd.read_excel('2019-11-09_LOCALIDADES_AFETADAS.xlsx')
```

```python
# Visualizando os dados.
df.tail(30)
```

```python
# Corrigindo o valor do índice 417.
df['localidade']=df['localidade'].replace('PORTO SAUIPE','Porto Sauípe')
```

```python
df.tail(30)
```

```python
# Selecionando as colunas do nosso interesse.
df = pd.DataFrame(df,columns=['localidade','municipio','Data_Avist','estado','Latitude','Longitude'])
```

```python
# Visualizando os dados depois das transformações.
df.head()
```

```python
# Verificando se existem valores duplicados.
df.duplicated()
```

```python
# Número de Localidades.
df['localidade'].count()
```

```python
# Número de Municípios.
df['municipio'].unique().shape[0]
```

```python
# Número de Estados.
df['estado'].nunique()
```

```python
# Analisando os tipos das colunas.
df.info()
```

```python
# Transformando a coluna 'Data_Avist' em datetime.
df['Data_Avist'] = df['Data_Avist'].astype('datetime64')
```

```python
# Verificando se o tipo da coluna foi alterado.
df.dtypes
```

```python
# Definindo a coluna 'Data_Avist' como índice do DataFrame.
df.set_index('Data_Avist',inplace=True)
```

```python
# Visualizando.
df.tail()
```

```python
# Salvando os dados.
df.to_csv('oleo_litoral.csv')
```
